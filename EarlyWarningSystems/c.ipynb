```python
import pandas as pd
```

```python
df_master = pd.read_excel('data/Alias Karyawan.xlsx')      
df_medical = pd.read_excel('data/Dummy Medical Data.xlsx')
```

```python
d_c = [
    "PAYOR_CODE", "PROVIDER NAME", "CLAIMS_ID", "CARD NO",
    "ADMISSION_DATE", "DISCHARGEABLE_DATE", "DUE_TOTAL",
    "APPROVE", "MEMBER PAY", "EXCESS PAID", "EXCESS NOT PAID", "STATUS"
]
```

```python
df_medical = df_medical.drop(columns=[col for col in d_c if col in df_medical.columns])
```

```python
for col in df_medical.columns:
    vc = df_medical[col].value_counts().reset_index()
    vc.columns = [col, 'count']
    vc.to_csv(f"{col}_value_counts.csv", index=False)
```

```python
print(df_medical.isnull().sum())
```

```python
df_result = df_medical.merge(
    df_master[['Name', 'Location']],  
    left_on='ALIAS_PATIENT',          
    right_on='Name',                 
    how='left'                       
)
```

```python
df_result.to_excel('data/merged_medical_with_location.xlsx', index=False)
```
