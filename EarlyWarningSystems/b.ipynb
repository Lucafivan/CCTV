```python
import pandas as pd
```

```python
df = pd.read_csv('data/DIAGNOSIS DESC_value_counts.csv')
```

```python
df.head()
```

```python
df.drop(columns="count", inplace=True)
```

```python
df.head()
```

```python
# Keyword accident / trauma
accident_keywords = [
    'fracture', 'injury', 'trauma', 'burn', 'laceration', 'wound', 'contusion', 'abrasion', 'bruise', 'impact', 'rupture',
    'dislocation', 'sprain', 'strain', 'tear', 'rupture of tendon', 'rupture of ligament', 'ligament injury', 'muscle injury',
    'fall', 'collision', 'traffic accident', 'motor vehicle accident', 'occupational injury', 'work accident',
    'open wound of head', 'closed fracture', 'clavicle fracture', 'finger fracture', 'ankle fracture', 'knee injury', 'head injury', 'cranial trauma'
]

# Tambahkan kolom baru
df['ACCIDENT_FLAG'] = 0

# Loop tiap baris
for i in range(len(df)):
    diagnosis = str(df.loc[i, 'DIAGNOSIS DESC']).lower()  # ambil dan lowercase
    for keyword in accident_keywords:
        if keyword in diagnosis:
            df.loc[i, 'ACCIDENT_FLAG'] = 1
            break  # keluar loop keyword kalau sudah match

# Lihat hasil
print(df)
```

```python
print(df[df["ACCIDENT_FLAG"] == 1])
```

Berikut adalah penjelasan lebih rinci mengenai 14 diagnosis yang tercatat dalam data Anda, lengkap dengan penyebab umum dan referensi medis terpercaya:

---

### 1. **Muscle Strain (Cedera Otot)**

**Penyebab Umum:**

* Aktivitas fisik berlebihan atau penggunaan otot secara tidak tepat.
* Cedera akibat kecelakaan atau jatuh.
* Penggunaan otot secara berulang tanpa pemanasan yang cukup.

**Referensi:**

* Cleveland Clinic: [Muscle Strains: Causes, Symptoms, Treatment & Recovery](https://my.clevelandclinic.org/health/diseases/22336-muscle-strains)

---

### 2. **Sprain and Strain involving Fibular (Tibial) Collateral Ligament of Knee (Cedera Ligamen Lutut Bagian Samping)**

**Penyebab Umum:**

* Cedera akibat putaran atau benturan pada lutut.
* Aktivitas olahraga yang melibatkan perubahan arah cepat.
* Trauma langsung pada lutut.

**Referensi:**

* OrthoInfo - AAOS: [Collateral Ligament Injuries](https://orthoinfo.aaos.org/en/diseases--conditions/collateral-ligament-injuries/)

---

### 3. **Fracture of Clavicle (Patah Tulang Selangka)**

**Penyebab Umum:**

* Jatuh langsung ke bahu atau tangan yang terulur.
* Cedera akibat benturan langsung pada bahu.
* Kecelakaan lalu lintas atau olahraga kontak.

**Referensi:**

* Mayo Clinic: [Broken Collarbone - Symptoms and Causes](https://www.mayoclinic.org/diseases-conditions/broken-collarbone/symptoms-causes/syc-20370311)

---

### 4. **Injury of Unspecified Body Region (Cedera pada Bagian Tubuh yang Tidak Ditentukan)**

**Penyebab Umum:**

* Cedera akibat kecelakaan, jatuh, aktivitas olahraga, atau kekerasan.
* Fraktur, memar, luka, atau cedera jaringan lunak lainnya.

**Referensi:**

* AAPC: [ICD-10 Code for Other Injury of Unspecified Body Region](https://www.aapc.com/codes/icd-10-codes/T14.8)

---

### 5. **Sprain and Strain of Ankle (Cedera Ligamen dan Otot Pergelangan Kaki)**

**Penyebab Umum:**

* Jatuh yang menyebabkan pergelangan kaki terpelintir.
* Mendarat tidak tepat setelah melompat atau berputar.
* Berjalan atau berolahraga di permukaan yang tidak rata.

**Referensi:**

* Mayo Clinic: [Sprained Ankle - Symptoms and Causes](https://www.mayoclinic.org/diseases-conditions/sprained-ankle/symptoms-causes/syc-20353225)

---

### 6. **Impacted Cerumen (Kotoran di Telinga)**

**Penyebab Umum:**

* Produksi kotoran telinga yang berlebihan.
* Penggunaan cotton bud atau alat pembersih telinga lainnya.
* Penggunaan alat bantu dengar atau earplug secara rutin.

**Referensi:**

* Cleveland Clinic: [Earwax Blockage: Symptoms, Causes & Removal](https://my.clevelandclinic.org/health/diseases/14428-ear-wax-buildup--blockage)

---

### 7. **Open Wound of Head (Luka Terbuka pada Kepala)**

**Penyebab Umum:**

* Cedera akibat benturan keras pada kepala.
* Kecelakaan lalu lintas, jatuh, atau kecelakaan olahraga.
* Luka akibat benda tajam atau tumpul.

**Referensi:**

* Cleveland Clinic: [Head Injury: Types, Symptoms, Causes & Treatments](https://my.clevelandclinic.org/health/diseases/head-injury)

---

### 8. **Impacted Teeth (Gigi Terdorong ke Dalam)**

**Penyebab Umum:**

* Gigi yang tidak tumbuh dengan posisi normal.
* Trauma atau benturan pada area mulut.
* Infeksi atau peradangan pada gusi.

**Referensi:**

* Mayo Clinic: [Impacted Wisdom Teeth - Symptoms and Causes](https://www.mayoclinic.org/diseases-conditions/wisdom-teeth/symptoms-causes/syc-20373808)

---

### 9. **Dislocation of Patella (Dislokasi Tempurung Lutut)**

**Penyebab Umum:**

* Cedera akibat benturan langsung pada lutut.
* Gerakan putaran atau perubahan arah yang tiba-tiba.
* Aktivitas olahraga yang melibatkan lompatan atau perubahan arah cepat.

**Referensi:**

* OrthoInfo - AAOS: [Patellar Dislocation](https://orthoinfo.aaos.org/en/diseases--conditions/patellar-dislocation)

---

### 10. **Fracture of Other Finger (Patah Jari Tangan Lainnya)**

**Penyebab Umum:**

* Jari terjepit atau terbentur keras.
* Cedera akibat aktivitas olahraga atau pekerjaan.
* Kecelakaan rumah tangga atau kendaraan.

**Referensi:**

* OrthoInfo - AAOS: [Hand Fractures](https://orthoinfo.aaos.org/en/diseases--conditions/hand-fractures)

---

### 11. **Sprain and Strain involving Anterior (Posterior) Cruciate Ligament of Knee (Cedera Ligamen Lutut Bagian Depan atau Belakang)**

**Penyebab Umum:**

* Cedera ligamen lutut (ACL/PCL) akibat gerakan tiba-tiba atau trauma olahraga.
* Aktivitas olahraga yang melibatkan perubahan arah cepat atau lompatan.
* Kecelakaan atau benturan langsung pada lutut.

**Referensi:**

* UPMC: [MCL/LCL Sprain Symptoms, Diagnosis, and Treatment](https://www.upmc.com/services/orthopaedics/conditions/mcl-lcl)

---

### 12. **Injuries to the Knee and Lower Leg (Contusion) (Cedera Memar pada Lutut dan Kaki Bawah)**

**Penyebab Umum:**

* Benturan langsung pada lutut atau kaki bawah.
* Jatuh atau kecelakaan yang menyebabkan cedera.
* Aktivitas olahraga yang melibatkan kontak fisik.

**Referensi:**

* UPMC: [Sprains and Strains Causes, Symptoms, and Treatments](https://www.upmc.com/services/orthopaedics/conditions/sprains-strains)

---

### 13. **Dislocation, Sprain and Strain of Joint and Ligaments at Ankle and Foot Level (Cedera Ligamen dan Sendi pada Pergelangan Kaki dan Kaki)**

**Penyebab Umum:**

* Jatuh atau mendarat tidak tepat setelah melompat.
* Berjalan atau berolahraga di permukaan yang tidak rata.
* Aktivitas olahraga yang melibatkan perubahan arah cepat.

**Referensi:**

* Mercy Health: [Ankle Sprains and Strains | Orthopedics](https://www.mercy.com/health-care-services/orthopedics-sports-medicine-spine/specialties/foot-ankle/conditions/ankle-sprains-strains)

---

### 14. **Unspecified Injury of Head (Cedera Kepala yang Tidak Ditentukan)**

**Penyebab Umum:**

* Cedera akibat benturan keras pada kepala.
* Kecelakaan lalu lintas, jatuh, atau kecelakaan olahraga.
* Luka akibat benda tajam atau tumpul.

**Referensi:**

* Cleveland Clinic: [Head Injury: Types, Symptoms, Causes & Treatments](https://my.clevelandclinic.org/health/diseases/head-injury)

| Top Risiko                                                | Masalah                                                                    | Solusi / Preventive Action                                                                                                                                          |
| --------------------------------------------------------- | -------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| 1. Kurangnya pemanasan / penggunaan otot tanpa persiapan  | Employee terbiasa bergerak tanpa pemanasan → risiko muscle strain / sprain | Edukasi rutin tentang **pemanasan / stretching** sebelum aktivitas fisik                                                                                            |
| 2. Aktivitas di area berisiko                             | Aktivitas di lokasi rawan cedera → jatuh, benturan                         | **Batasi akses** area rawan, buat **aturan tambahan / signage**, monitor aktivitas                                                                                  |
| 3. Kurangnya penggunaan alat pelindung / safety equipment | Cedera akibat tidak memakai alat pelindung → kepala, kaki, tangan          | Pastikan **helm, sarung tangan, sepatu safety, ADP** digunakan sesuai standar; upgrade seragam untuk proteksi tambahan                                              |
| 4. Masalah pada telinga / paparan suara                   | Risiko impacted cerumen atau gangguan telinga karena paparan suara tinggi  | Edukasi **cara membersihkan telinga yang benar**, gunakan **headset / peredam suara**, pasang **mesin pendeteksi desibel** untuk monitoring area dan waktu tertentu |
| 5. Luka akibat benda tajam / tumpul                       | Cedera fisik dari benda tajam atau benturan                                | Gunakan **ADP khusus**, upgrade seragam, pastikan **area kerja bebas risiko** dari benda berbahaya                                                                  |
